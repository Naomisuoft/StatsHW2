Q.6 The df.describe() method provides the 'count', 'mean', 'std', 'min', '25%', '50%', '75%', and 'max' summary statistics for each variable it analyzes. Give the definitions (perhaps using help from the ChatBot if needed) of each of these summary statistics
  Each of these values are summary statistics used to find the distibution of values in a data set.
Count shows the amount of existing values in each numerical column. The mean is the average value of the values in a column, calculated by dividing the sum of all the existing values in the column by the count of the values. STD is the standard deviation of the data points, meaning the dispersion of the data points in reference to the mean.
The max and min are the highest and lowest value in each column representing numerical data. 25%, 50% and 75% are quartiles into which the data points can be organized. The first quartile (25%) is the point before which a quarter of the data points lie, the median (50%) is the middle point of the data when all the values are organized from smallest to largest, finally, the third quartile (75%), is the point before which 75% of the data points lie.
